```python
# Imports
import tensorflow as tf
from tensorflow.keras.utils import Sequence
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input, Dropout
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
file = str('target.dat')
f = open(file,'r')
linea = f.readline()
fields = linea.split()
dim = int(fields[0])
target = np.ndarray(shape=(dim))
print(target.shape)
for i in range(dim):
    linea = f.readline()
    fields = linea.split()
    target[i] = float(fields[0])
f.close()
```

```python
file = str('data.dat')
f = open(file,'r')
linea = f.readline()
fields = linea.split()
dim = int(fields[0])
data = np.ndarray(shape=(dim,9))
print(data.shape)
for i in range(dim):
    linea = f.readline()
    fields = linea.split()
    for j in range(9):
        data[i][j] = float(fields[j])
f.close()
```

```python
ked_input = np.ndarray(shape=(target.shape[0],5))
for i in range(target.shape[0]):
    ked_input[i][0] = np.exp(-0.20*np.power(data[i][0],2))
    for j in range(4):
        ked_input[i][j+1]=data[i][j+1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(ked_input,target,test_size=0.50)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
def KED_model(input_shape = (5,)):
    
    inputs = Input(shape=input_shape)
    x = Dense(100,activation='relu')(inputs)
    x = Dense(100,activation='relu')(x)
    x = Dense(100,activation='relu')(x)
    x = Dense(100,activation='relu')(x)
    x = Dense(100,activation='relu')(x)
    x = Dense(1,activation='sigmoid')(x)
    KED = tf.keras.Model(inputs=inputs, outputs= x)
    
    return KED
```

```python
model = KED_model()
model.summary()
```

```python
model.compile(optimizer='adam',loss='mape',metrics=['mae'])
```

```python
history = model.fit(x_train,y_train,validation_split=0.5,epochs=10000,batch_size=256)
```

```python
model.save("augKED_model.h5",overwrite=True)
```

```python
test_mape_score,test_mae_score = model.evaluate(x_test,y_test)
```
